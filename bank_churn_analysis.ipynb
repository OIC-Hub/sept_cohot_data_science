```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load dataset
df1 = pd.read_excel("Bank_Churn_Messy.xlsx", sheet_name=0)
df2 = pd.read_excel("Bank_Churn_Messy.xlsx", sheet_name=1)

df = pd.merge(df1, df2, on=["CustomerId", "Tenure"])

df.head(20)
```

```python
#inspect the dataset
df.info()
```

```python
#check for missing values
df.isnull().sum()
```

```python
#check for duplicate
duplicated = df.duplicated().sum()
print(duplicated)
```

```python
#handle missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#handle duplicated values
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.head(20)
```

```python
df["Geography"].unique()
```

```python
df["Geography"].replace({
    "FRA": "France",
    "French": "France"
}, inplace=True)

df["Geography"].unique()
```

```python
def clean_currency(x):
    if isinstance(x, str):
        return x.replace("€", "")
    return x
df["EstimatedSalary"] = df["EstimatedSalary"].apply(clean_currency)
df["Balance"] = df["Balance"].apply(clean_currency)
df.head()
```

```python
#convert exited column with (1 - yes, 0 - No)
df["Exited"] = df["Exited"].replace({0: "No", 1: "Yes"})
df.head()
```

```python
# select customer with highest credit score
high_customer_credit_score = df[df["CreditScore"] >= 700]
high_customer_credit_score
```

```python
#find the active customer in germany
active_customer_in_germany = df[(df["Geography"] == "Germany") & (df["IsActiveMember"] == "Yes")]
active_customer_in_germany
```

```python
churned_customer = df[df["Exited"] == "Yes"]
churned_customer
```

```python
# create a costume column called "servicerating"
np.random.seed(42)
ratings = pd.DataFrame({
    "CustomerId": df["CustomerId"].unique(),
    "ServiceRating": np.random.randint(1, 6, len(df["CustomerId"].unique()))
})

df = df.merge(ratings, on="CustomerId")
df
```

```python
avg_credit_score = df.groupby("Geography")["CreditScore"].mean()
avg_credit_score
```
